```python
import pandas as pd

df = pd.read_csv('student_data.csv')
```

```python
df.groupby('placement_status')['placement_status'].agg('count')
```

```python
df.groupby('placement_status')['exam_score'].agg('std')
```

```python
df.groupby('placement_status')['study_hours'].agg('mean')
```

```python
df.groupby('study_hours')['exam_score'].agg('mean')
```
